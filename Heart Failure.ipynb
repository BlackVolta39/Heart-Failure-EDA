# Outlier Treatment

```python
#Decide weather to do outlier treatment or not, how do you do it?
def wisker(col):
    q1,q3 = np.percentile(col,[25,75])
    iqr = q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
wisker(df["serum_creatinine"])
```

```python
for i in ['serum_creatinine','serum_sodium','platelets','creatinine_phosphokinase','ejection_fraction']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
#Checking if your outliers have been resolved
for i in ['serum_creatinine','serum_sodium','platelets','creatinine_phosphokinase','ejection_fraction']:
    sns.boxplot(df[i])
    plt.show()
```
